```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
yt = np.linspace(0, 10, 100)
```

```python
N = 5
```

```python
n0 = 3
delta_n = 10
k = 3
c = 1e-3
```

```python
f = lambda i: np.sqrt(delta_n)*2**k*c*(n0/(i*delta_n))*np.cos(2*np.pi*i*delta_n*yt + np.random.randn())
```

```python
harmonics = []
for j in range(1, N):
    harmonics.append(f(j))
```

```python
for h in harmonics:
    plt.plot(h)
```

```python
plt.plot(np.sum(np.array(harmonics), axis=0))
```
