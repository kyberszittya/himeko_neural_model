```python
def colorize(z):
    n,m = z.shape
    c = np.zeros((n,m,3))
    c[np.isinf(z)] = (1.0, 1.0, 1.0)
    c[np.isnan(z)] = (0.5, 0.5, 0.5)

    idx = ~(np.isinf(z) + np.isnan(z))
    A = (np.angle(z[idx]) + np.pi) / (2*np.pi)
    A = (A + 0.5) % 1.0
    B = 1.0 - 1.0/(1.0+abs(z[idx])**0.3)
    c[idx] = [hls_to_rgb(a, b, 0.8) for a,b in zip(A,B)]
    return c
```

## Generating road profile data

```python
N = 6
delta_gen = 0.1
gamma_k = lambda k: (2*np.pi * k) / (N * delta_gen)
```

```python
def iso_spectral_density(kappa, _kappa0, suk0):
    a = kappa / _kappa0
    __n1 = 3
    __n2 = 2.25
    return suk0 * np.where(a <= 1, a ** -__n1, a ** -__n2)
```

```python
freqs = np.fft.fftfreq(250)
kappa0 = 1/(2*np.pi)
iso_target = np.array([2, 8, 32, 128, 512, 2048])*1e-6
plt.ylim([-1,1])
plt.grid()
for i in range(len(iso_target)):
    plt.plot(freqs, iso_spectral_density(freqs, kappa0, iso_target[i]))
```

```python
def profile_generation(L, target):
    N = len(L)

    denom =  (1/N)*(1/delta_gen)
    
    gen_prof = []
    vals = []
    for t in L:
        phi_k = np.random.uniform(0, 2*np.pi, N)
        v_t = []
        v0 = 0.0
        for k in range(1, N + 1):
            gk = (2*np.pi * k) * denom
            S11 = iso_spectral_density(gk, kappa0, suk0=target)
            Sk = S11 * (2 * np.pi) * denom
            #v = np.sqrt(Sk)*np.exp(1j*(phi_k[k] + (2*np.pi * k * r)/N))
            #phi_k = np.random.uniform(0, 2*np.pi)
            v = np.sqrt(Sk)*np.exp(1j*(phi_k[k-1] + (2*np.pi * k * t)/N))
            v_t.append(v)
            v0 += v
        gen_prof.append(v_t)
        vals.append(v0)
    return np.array(gen_prof), np.array(vals)
```

```python
delta_gen = 0.1
L = np.arange(0, 50, delta_gen)

profile_A, vals = profile_generation(L, iso_target[0])
plt.plot(vals)
plt.show()
```

## White noise based road profile

```python
def white_noise_spectral_density(wavenumber, ga, gs, ge):    
    return ga/(2*np.pi*wavenumber)**4 + gs/(2*np.pi*wavenumber)**2 + ge 
```

```python
print(white_noise_spectral_density(1.0, 4, 6, 7))
```

```python
freqs = np.fft.fftfreq(250)
kappa0 = 1/(2*np.pi)
iso_target = np.array([2, 8, 32, 128, 512, 2048])*1e-6
plt.ylim([-0.1,1.5])
plt.xlim([-0.1,0.1])
plt.grid()


ra = np.array(range(1,6))
gs = ra*50*1e-6
ga = ra*1e-6
ge = ra*1e-6

for i in range(len(ra)):
    v =  white_noise_spectral_density(freqs, ga[i], gs[i], ge[i])
    plt.plot(freqs, v)
```

```python
def profile_generation_white_noise(L, gs, ga, ge):
    N = len(L)

    denom =  (1/N)*(1/delta_gen)

    gen_prof = []
    vals = []
    for t in L:
        phi_k = np.random.uniform(0, 2*np.pi, N)
        v_t = []
        v0 = 0.0
        for k in range(1, N):
            gk = (2*np.pi * k) * denom
            S11 =  white_noise_spectral_density(gk, ga, gs, ge)
            Sk = S11 * (2 * np.pi) * denom
            #v = np.sqrt(Sk)*np.exp(1j*(phi_k[k] + (2*np.pi * k * r)/N))
            phi_k = np.random.uniform(0, 2*np.pi)
            v = np.sqrt(Sk)*np.exp(1j*(phi_k + (2*np.pi * k * t)/N))  
            v_t.append(v)
            v0 += v
        gen_prof.append(v_t)
        vals.append(v0)
    return np.array(gen_prof), np.array(vals)
```

```python
def generate_normal_profile(gs, ga, ge, delta_gen, n):
    sigma_gs = (np.mean(gs))/(2*delta_gen)
    gen_gs = np.random.normal(loc=0.0, scale=sigma_gs, size=(n,))
    sigma_ga = (np.mean(ga))/(2*delta_gen)
    gen_ga = np.random.normal(loc=0.0, scale=sigma_ga, size=(n,))
    sigma_ge = (np.mean(ge))/(2*delta_gen)
    gen_ge = np.random.normal(loc=0.0, scale=sigma_ge, size=(n,))
    return gen_gs, gen_ga, gen_ge
    
```

## Robson's spectrum

```python
def su_robson(s0, wavenumber, n):
    return s0*np.abs(wavenumber)**-n
```

```python
v = su_robson(50e-8, np.fft.fftfreq(100, 0.1), 2.5)
```

```python
plt.plot(v)
```

```python
def profile_generation_robson(L, gs, ga, ge):
    N = len(L)

    denom =  (1/N)*(1/delta_gen)

    gen_prof = []
    vals = []
    for t in L:
        phi_k = np.random.uniform(0, 2*np.pi, N)
        v_t = []
        v0 = 0.0
        for k in range(1, N):
            gk = (2*np.pi * k) * denom
            S11 =  white_noise_spectral_density(gk, ga, gs, ge)
            Sk = S11 * (2 * np.pi) * denom
            #v = np.sqrt(Sk)*np.exp(1j*(phi_k[k] + (2*np.pi * k * r)/N))
            phi_k = np.random.uniform(0, 2*np.pi)
            v = np.sqrt(Sk)*np.exp(1j*(phi_k + (2*np.pi * k * t)/N))
            v_t.append(v)
            v0 += v
        gen_prof.append(v_t)
        vals.append(v0)
    return np.array(gen_prof), np.array(vals)
```

```python
def complex_profile_generation(_spati, _x, dx, L, Gdn0=2e-6):
    _components = int(L/dx/2)
    phase = np.random.uniform(0, 2*np.pi, _components)
    vals_complex = []
    kappa0 = 1/(2*np.pi)
    gamma = 2*np.pi*_spati/_components*dx
    Gd_c = np.sqrt(Gdn0 * (2*np.pi/_components*dx) * ((gamma/2)/kappa0)**(-2))
    for i in range(len(_x)):
        v = np.sum(Gd * np.cos(2*np.pi*_spati*_x[i] + phase))
        #v_complex = np.sum(Gd_c * np.exp(1j*2*np.pi*spati*x[i] + phase))
        v_complex = np.sum(Gd_c * (np.exp(1j*((2*np.pi*_spati*_x[i]) + phase))/_components).real)
        vals_complex.append(v_complex)
    return np.array(vals_complex)

dx = 0.1
L = 500
components = int(L/dx/2)

Gdn0 = 2e-6
x = np.arange(0, L+dx/2., dx)
n0 = 0.1

n_min = 0.0078
n_max = 4.0
spati = np.linspace(n_max, n_min, components)

vals_complex = complex_profile_generation(spati, x, dx, L, 2e-6)

plt.plot(x, vals_complex)
plt.show()
```

```python
profiles = []
for gdn in [2e-6, 8e-6, 32e-6, 128e-6, 512e-6, 2048e-6]:
    vals_complex = complex_profile_generation(spati, x, dx, L, gdn)
    profiles.append(vals_complex)
    plt.plot(x, vals_complex)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(14,20))
for vals_complex in profiles:
    profile_fft = np.fft.fft(vals_complex)
    ps = np.abs(profile_fft ** 2)
    len_data = int(ps.shape[0]/2)
    freqs = np.fft.fftfreq(len_data, dx)
    nyquist_freq = freqs[:len_data]
    plt.loglog(ps[:len_data]/(2*np.pi)/len_data)
plt.grid(which='minor')
```
