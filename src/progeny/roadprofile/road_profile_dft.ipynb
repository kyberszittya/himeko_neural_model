```python
from colorsys import hls_to_rgb
```

```python
def colorize(z):
    n,m = z.shape
    c = np.zeros((n,m,3))
    c[np.isinf(z)] = (1.0, 1.0, 1.0)
    c[np.isnan(z)] = (0.5, 0.5, 0.5)

    idx = ~(np.isinf(z) + np.isnan(z))
    A = (np.angle(z[idx]) + np.pi) / (2*np.pi)
    A = (A + 0.5) % 1.0
    B = 1.0 - 1.0/(1.0+abs(z[idx])**0.3)
    c[idx] = [hls_to_rgb(a, b, 0.8) for a,b in zip(A,B)]
    return c
```

```python
def road_quality_spectra(k, kappa, kappa0):
    a = kappa/kappa0
    return k*1e-6*np.where(a <= 1, a**-3, a**-2.25)
```

```python
plt.yscale('log', base=10)
plt.grid()
plt.ylim([1e-8, 1e-2])
plt.loglog(T, road_quality_spectra(2, T, kappa0), color='violet', linestyle='dashed', label='very good')
plt.loglog(T, road_quality_spectra(8, T, kappa0), color='blue', linestyle='dashed', label='good')
plt.loglog(T, road_quality_spectra(32, T, kappa0), color='green', linestyle='dashed', label='average')
plt.loglog(T, road_quality_spectra(128, T, kappa0), color='orange', linestyle='dashed', label='poor')
plt.loglog(T, road_quality_spectra(512, T, kappa0), color='red', linestyle='dashed', label='very poor')
plt.loglog(T, road_quality_spectra(2048, T, kappa0), color='brown', linestyle='dashed', label='atrocious')
```

```python
def plot_profile_quality(freq):
    plt.grid()
    plt.ylim([1e-8, 1e-2])
    plt.loglog(freq, road_quality_spectra(2, freq, kappa0), color='violet', linestyle='dashed', label='very good')
    plt.loglog(freq, road_quality_spectra(8, freq, kappa0), color='blue', linestyle='dashed', label='good')
    plt.loglog(freq, road_quality_spectra(32, freq, kappa0), color='green', linestyle='dashed', label='average')
    plt.loglog(freq, road_quality_spectra(128, freq, kappa0), color='orange', linestyle='dashed', label='poor')
    plt.loglog(freq, road_quality_spectra(512, freq, kappa0), color='red', linestyle='dashed', label='very poor')
    plt.loglog(freq, road_quality_spectra(2048, freq, kappa0), color='brown', linestyle='dashed', label='atrocious')
```

```python
plot_profile_quality(nyquist_freq)
```
