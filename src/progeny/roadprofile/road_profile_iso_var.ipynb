```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sampling interval
B = 0.4
L = 250
nmax = 1.0/B

deltaN = 1 / L
N = int(nmax / deltaN)
n0 = 0.1
```

```python
phi = np.random.uniform(0, 2.0*np.pi, N)
plt.plot(phi)
```

```python
def profile(x, N, k, deltaN, n0):
    h = []
    for i in range(1,N):
        phi = np.random.uniform(0, 2.0*np.pi)
        h.append(np.sqrt(deltaN)*(2**k)*0.001*(n0/(i*deltaN)) * np.cos((2*np.pi*i*deltaN*x) + phi))
    
    return np.sum(np.array(h))
```

```python
def generate_profile(k, t, N, deltaN, n0):
    vals = []    
    for x in t:
        vals.append(profile(x, N, k, deltaN, n0))
    return np.array(vals)
    
```

```python
t = np.arange(0, L, B)
vals_ISO_A = generate_profile(3, t, 6, deltaN, n0)
plt.figure(figsize=(25,14))
plt.plot(t, vals_ISO_A, label='ISO A')
plt.legend()
plt.grid()
```
